```python
## Square root of num by use map:

def sqr(r):
    return r*r

sqr(99)
```

```python
num=[1,2,39,56,73]
list(map(sqr,num))
```

```python
##Adding multiple parameters in map:

val1=[36,78,45,26]
val2=[43,71,93,42]

add=list(map(lambda a,b:a+b,val1,val2))
print(add)
```

```python
##onverting small letter to capital letter using map:

smallcap=['rohan','sharma','krish','naik']
capital=list(map(str.upper,smallcap))
print(capital)
```

```python
## extract name from the data:

def ext_name(person):
    return person['name']
data=[{'name':'rohan','age':23},
      {'name':'krish','age':34},
     {'name':'Aishwarya','age':28}]
list(map(ext_name,data))
capital=list(map(str.upper,list(map(ext_name,data))))
print(capital)
    
                
```
